```python
import pandas as pd

# Create a simple dataset with special characters, uppercase letters, and numbers
data = {'text': ["Free money now!!! Click this link: http://example.com",
                   "Hi, how are you today?",
                   "Claim your $1,000,000 prize NOW!!!",
                   "Meeting tomorrow at 10 AM sharp.",
                   "URGENT: Your account has been compromised. Click here: https://secure.site/login",
                   "Just wanted to follow up on our conversation regarding Project X.",
                   "Congratulations! You've won a FREE vacation to Hawaii!",
                   "Looking forward to seeing you at the event #BestEvent"],
        'label': ["spam",
                  "not spam",
                  "spam",
                  "not spam",
                  "spam",
                  "not spam",
                  "spam",
                  "not spam"]}

df_spam = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df_spam.to_csv('spam.csv', index=False)

print("'spam.csv' created successfully with special characters, uppercase letters, and numbers!")
```

```python
display(df_spam.head())
```

```python
import re
import string

def clean_text(text):
    text = text.lower() # Convert to lowercase
    text = re.sub(f'[{re.escape(string.punctuation)}]', '', text) # Remove punctuation
    text = re.sub(r'\d+', '', text) # Remove numbers
    text = text.strip() # Remove leading/trailing whitespace
    text = re.sub(r'\s+', ' ', text) # Remove extra spaces
    return text

df_spam['cleaned_text'] = df_spam['text'].apply(clean_text)

print("Text cleaning complete. Added 'cleaned_text' column.")
display(df_spam[['text', 'cleaned_text']].head())
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Download the missing resource
```

```python
from nltk.tokenize import word_tokenize

df_spam['tokens'] = df_spam['cleaned_text'].apply(word_tokenize)

print("Tokenization complete. Added 'tokens' column.")
display(df_spam[['cleaned_text', 'tokens']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF Vectorizer
# Using cleaned_text as the input for TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=1000) # You can adjust max_features

# Fit and transform the cleaned text data
tfidf_features = tfidf_vectorizer.fit_transform(df_spam['cleaned_text'])

print("TF-IDF Feature Extraction complete.")
print("Shape of TF-IDF features:", tfidf_features.shape)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(tfidf_features, df_spam['label'], test_size=0.2, random_state=42) # You can adjust test_size and random_state

print("Data splitting complete.")
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Initialize a Logistic Regression model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)

print("Model training complete.")

# Make predictions on the test set
y_pred = model.predict(X_test)

print("Predictions made on the test set.")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:")
print(report)
```
